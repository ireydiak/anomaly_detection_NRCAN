```python
import pandas as pd
# importing package
import matplotlib.pyplot as plt
import numpy as np
dagmm_misclf_df = pd.read_csv("05-03-2022_12-32-25/dagmm_misclassifications.csv", index_col=0)
dagmm_misclf_df = dagmm_misclf_df.sort_index()
deepsvdd_misclf_df = pd.read_csv("deepsvdd_experiments/deepsvdd_misclassifications.csv", index_col=0)
deepsvdd_misclf_df = deepsvdd_misclf_df.sort_index()
```

```python
# create data
x = np.arange(len(dagmm_misclf_df.index))
y1 = 1 - dagmm_misclf_df["Misclassified ratio"]
y2 = 1 - deepsvdd_misclf_df["Misclassified ratio"]
width = 0.2

f, ax = plt.subplots(figsize=(15, 8))
# plot data in grouped manner of bar type
ax.bar(x - 0.2, y1, width, color='blue')
ax.bar(x, y2, width, color='orange')
plt.xticks(x, dagmm_misclf_df.index)
plt.xlabel("Labels")
plt.ylabel("Good classification rate")
plt.legend(["DAGMM", "DeepSVDD"])
plt.show()
```
